## Map of California

This map of California shows where all the data points are.

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

# Create a Cartopy map with PlateCarree projection (lat/lon)
fig = plt.figure(figsize=(10, 10))
ax = plt.axes(projection=ccrs.PlateCarree())

# Set map extent: West Coast (California)
ax.set_extent([-125, -114, 32, 43])  # [west, east, south, north]

# Add base map features
ax.coastlines(resolution='10m')
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.STATES, linestyle='--')
ax.gridlines(draw_labels=True)

# Plot pier locations from df_cal_habs
ax.scatter(
    df_cal_habs["longitude"],
    df_cal_habs["latitude"],
    color="red",
    s=50,
    marker="^",
    transform=ccrs.PlateCarree(),
    label="California Piers"
)

# Optional: Add text labels using the Location_Code column
for _, row in df_cal_habs.iterrows():
    ax.text(
        row["longitude"] + 0.05,  # Offset label slightly to the right
        row["latitude"],
        row["Location_Code"],
        fontsize=9,
        transform=ccrs.PlateCarree()
    )

# Final touches
plt.title("California Pier Locations from df_cal_habs")
plt.legend()
plt.show()
```
